## [DataFrame 정렬]

- 종류 => 인덱스 기반 정렬, 데이터/값 기반 정렬

(1) 모듈 로딩 <hr>

```python
import pandas as pd
```

(2) 데이터 준비 <hr>

```python
datafile = '../DATA/movies.csv'
```

(3) 데이터 저장 => DataFrame
- 파일 데이터 읽어서 저장 => pandas.read_파일확장자() ==> 반환 : DataFrame

```python
movieDF = pd.read_csv(datafile)
```

(4) 데이터 확인

```python
movieDF
```

- 데이터 확인 관련 메서드

```python
movieDF.info()
```

```python
# 컬럼별 통계적인 데이터 정보 제공
twoDF = movieDF.describe(include='all')
twoDF
```

```python
# 수치 컬럼의 최소, 최대, 평균 데이터 추출
twoDF.loc[['min', 'max', 'mean']]
```

- #### 행=>열, 열=>행 치환 기능 속성 및 메서드

```python
pd.DataFrame([twoDF.Rank.values, twoDF.Year.values], index=twoDF.columns, columns=twoDF.index)
```

- 속성 : DF객체변수명.T => 원본 데이터 변경 안됨!

```python
tt = twoDF.T
```

```python
tt
```

- 메서드 : DF객체변수명.transpose()

```python
twoDF = twoDF.transpose()
```

```python
twoDF
```

(5) 데이터 정렬
- 인덱스 기반 정렬 => 행인덱스열인덱세

```python
# 행 인덱스 기반 정렬 => DF.sort_index()
movieDF.sort_index(ascending=False)
```

```python
# 컬럼 인덱스 기반 정렬 => 내림차순
movieDF.sort_index(axis='columns', ascending=False)
```

(5) 데이터 정렬
- 데이터/값 기반 정렬

```python
movieDF.sort_values(by='Year')
```

```python
movieDF.sort_values(by=['Year', 'Title'])
```
