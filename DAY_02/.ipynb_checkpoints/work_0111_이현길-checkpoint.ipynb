#### [과제 및 복습]

- csv 데이터파일 읽어와서 Series, DataFrame 속성 읽기, 속성 변경
- 원소 읽기, 원소값 변경

(1) 모듈 로딩

```python
from pandas import DataFrame, read_csv
```

(2) 데이터 준비

```python
dataFile = '../DATA/movies.csv'
```

(3) 데이터 저장

```python
movieDF = read_csv(dataFile)
movieDF
```

(4) DataFrame의 속성들 출력

```python
print(movieDF.index, movieDF.columns, movieDF.values,
       movieDF.ndim, movieDF.shape, movieDF.size, movieDF.dtypes, sep='\n')
```

(5) DataFrame의 컬럼별로 따로 저장하기

```python
columnList = []
for col in movieDF.columns:
    columnList.append(movieDF[col])
```

(6) 컬럼별 데이터들의 속성 즉, index, values, shape. ndim, size, dtype 정보 출력

```python
for col in columnList:
    print(col.index, col.values, col.shape, col.ndim, col.size, col.dtype, sep='\n')
    print()
```

(7) 컬럼별 인덱스를 변경 실습

```python
for col in columnList:
    col.index = [f'No.{i:0>3}' for i in range(col.size)]
columnList[1]
```

(8) 컬럼별 데이터 즉 값 변경

```python
movieDF['Title'][0] = '어밴져스: 엔드게임'
movieDF['Title'][1] = '아바타'
```

(9) DataFrame의 원소들 모두 출력 단, 2가지 방법 사용

```python
for row in movieDF.index:
    for col in movieDF.columns:
        print(movieDF.loc[row, col])
```

```python
for row in range(movieDF.shape[0]):
    for col in range(movieDF.shape[1]):
        print(movieDF.iloc[row, col])
```
