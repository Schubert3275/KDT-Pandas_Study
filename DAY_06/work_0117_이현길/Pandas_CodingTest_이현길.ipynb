[문3] 아래 조건을 만족하는 데이터 객체를 생성하세요.
1) 데이터 : stats_100701.xlsx
2) 년도별 인구가 가장 많은 지역과 가장 적은 지역 
3) 6대 광역시에서 년도별 인구가 가장 많은 지역과 가장 적은 지역 
4) 8도 중 7도에서 년도별 인구가 가장 많은 지역과 가장 적은 지역 

```python
file2 = './stats_100701.xlsx'
```

```python
statsDF = pd.read_excel(file2)
statsDF = statsDF.set_index('지역별 인구  [단위 : 천명, 명/㎢]')
statsDF.columns = ['2015', '2016', '2017', '2018']
statsDF = statsDF.drop(np.nan)
```

```python
statsDF
```

2) 년도별 인구가 가장 많은 지역과 가장 적은 지역

```python
print("연도별 가장 인구가 많은 지역")
for year in statsDF.columns:
    print(f'{year}년 : {statsDF[year][1:-1].sort_values(ascending=False).head(1).index[0]}')
print("연도별 가장 인구가 적은 지역")
for year in statsDF.columns:
    print(f'{year}년 : {statsDF[year][1:-1].sort_values().head(1).index[0]}')
```

3) 6대 광역시에서 년도별 인구가 가장 많은 지역과 가장 적은 지역

```python
citys = ['서울', '부산', '대구', '인천', '광주', '대전']
print("6대 광역시에서 연도별 가장 인구가 많은 지역")
for year in statsDF.columns:
    print(f'{year}년 : {statsDF[year][citys].sort_values(ascending=False).head(1).index[0]}')
print("6대 광역시에서 연도별 가장 인구가 적은 지역")
for year in statsDF.columns:
    print(f'{year}년 : {statsDF[year][citys].sort_values().head(1).index[0]}')
```

4) 8도 중 7도에서 년도별 인구가 가장 많은 지역과 가장 적은 지역 

```python
provinces = ['경기', '충북', '충남', '전북', '전남', '경북', '경남']
print("7도에서 연도별 가장 인구가 많은 지역")
for year in statsDF.columns:
    print(f'{year}년 : {statsDF[year][provinces].sort_values(ascending=False).head(1).index[0]}')
print("7도에서 연도별 가장 인구가 적은 지역")
for year in statsDF.columns:
    print(f'{year}년 : {statsDF[year][provinces].sort_values().head(1).index[0]}')
```

[문4] 아래 조건을 만족하는 데이터 객체를 생성하세요.
1) 데이터 : stock price.xlsx
2) 데이터 로딩 시 인덱스로 사용할 수 있는 컬럼 설정
3) 결측치 처리 =>  
   (가) 1개라도 NaN이면 모두 제거 / price가 NaN이면 제거 / value가 10000이하면 제거  
   (나) 평균값으로 결측치 대체, 이전/이후 값으로 결측치 대체
4) value가 1000이하 이거나 price가 90000이상인 데이터

```python
file3 = './stock price.xlsx'
```

2) 데이터 로딩 시 인덱스로 사용할 수 있는 컬럼 설정

```python
stockDF = pd.read_excel(file3, index_col='id')
```

```python
stockDF
```

3) 결측치 처리 => 1개라도 NaN이면 모두 제거

```python
stockDF_any_NaN = stockDF.dropna()
stockDF_any_NaN
```

3) 결측치 처리 => price가 NaN이면 제거

```python
stockDF_price_NaN = stockDF.dropna(subset=['price'])
stockDF_price_NaN
```

3) 결측치 처리 => value가 10000이하면 제거

```python
stockDF_value_under = stockDF[stockDF['value'] > 10000]
stockDF_value_under
```

3) 결측치 처리 => 평균값으로 결측치 대체

```python
stockDF_mean_NaN = stockDF.fillna(stockDF.mean(numeric_only=True))
stockDF_mean_NaN
```

3) 결측치 처리 => 이전 값으로 결측치 대체

```python
stockDF_f_NaN = stockDF.fillna(method='ffill')
stockDF_f_NaN
```

3) 결측치 처리 => 이후 값으로 결측치 대체

```python
stockDF_b_NaN = stockDF.fillna(method='bfill')
stockDF_b_NaN
```

4) value가 1000이하 이거나 price가 90000이상인 데이터

```python
v_1000 = stockDF['value'] <= 1000
p_90000 = stockDF['price'] >= 90000
stockDF_or = stockDF[v_1000 | p_90000]
stockDF_or
```
