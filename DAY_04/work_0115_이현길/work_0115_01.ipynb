## [남북한 발전 전력량]

```python
import pandas as pd
import numpy as np
```

```python
file = './남북한발전전력량.xlsx'
```

```python
powers = pd.read_excel(file)
```

```python
powers
```

```python
powers.loc[9] = '-'
powers
```

```python
powers.loc[9, '전력량 (억㎾h)':'발전 전력별'] = (np.nan, '신재생')
```

```python
powers
```

```python
powers.info()
```

```python
for col in powers.columns[2:]:
    powers[col] = powers[col].replace('-', 0).astype(int)
```

```python
powers.info()
```

```python
powers
```

```python
for i in powers.index:
    powers.loc[i, '총합'] = powers.loc[i].values[2:-1].sum()
```
