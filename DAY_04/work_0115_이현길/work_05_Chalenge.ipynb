## [코딩 챌린지]

```python
import pandas as pd
```

```python
file = '../../DATA/netflix.csv'
```

1. 제한된 메모리 사용량과 사용성을 최대화하기 위해 데어터셋을 최적화하세요.

```python
netflixDF = pd.read_csv(file, parse_dates=['date_added'])
```

```python
netflixDF
```

```python
netflixDF.info()
```

```python
netflixDF.nunique()
```

```python
netflixDF.type = netflixDF.type.astype('category')
```

```python
netflixDF.info()
```

2. 제목이 'Limitless'인 모든 콘텐츠를 찾으세요.

```python
title_is_linitless = netflixDF.title == 'Limitless'
netflixDF[title_is_linitless]
```

3. 'Robert Rodriguez' 감독이 제작하고 유형이 'Movie'인 모든 콘텐츠를 찾으세요.

```python
director_is_rodriguez = netflixDF.director == 'Robert Rodriguez'
type_movie = netflixDF.type == 'Movie'
netflixDF[director_is_rodriguez & type_movie]
```

4. 등록된 날짜가 '2019-07-31'이거나 감독이 'Robert Altman'인 모든 콘텐츠를 찾으세요.

```python
registration_date = netflixDF.date_added == '2019-07-31'
director_is_altman = netflixDF.director == 'Robert Altman'
netflixDF[registration_date | director_is_altman]
```

5. 감독이 'Orson Welles', 'Aditya Kripalani' 또는 'Sam Raimi'인 모든 콘텐츠를 찾으세요.

```python
directors = ['Orson Welles', 'Aditya Kripalani', 'Sam Raimi']
find_director = netflixDF.director.isin(directors)
netflixDF[find_director]
```

6. 2019년 5월 1일과 2019년 6월 1일 사이에 등록된 모든 콘텐츠를 찾으세요.

```python
between_date = netflixDF.date_added.between('2019-05-01', '2019-06-01')
netflixDF[between_date]
```

7. director 열에서 NaN값이 있는 모든 행을 삭제하세요.

```python
netflixDF.dropna(subset=['director'])
```

8. 넷플릭스가 콘텐츠를 단 하나만 등록한 날짜를 찾으세요.

```python
netflixDF.drop_duplicates(subset=['date_added'], keep=False)
```
