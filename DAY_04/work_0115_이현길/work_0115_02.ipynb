## [시도별 전출입 인구수]

```python
import pandas as pd
import numpy as nb
```

```python
file = './시도별 전출입 인구수.xlsx'
```

```python
peoples = pd.read_excel(file)
```

```python
peoples
```

```python
peoples['전출지별'].fillna(method='ffill', inplace=True)
```

```python
peoples
```

```python
peoples = peoples.drop(0)
```

```python
for col in peoples.columns[2:]:
    peoples[col] = peoples[col].replace('-', 0).astype(int)
peoples.info()
```

```python
peoples2 = peoples.set_index(['전출지별', '전입지별'])
peoples2
```

```python
peoples2.index
```

```python
peoples3 = peoples2.swaplevel('전출지별', '전입지별')
peoples3
```

```python
peoples3.loc['전국']
```

```python
# for idx in peoples2.loc['전국'].index:
    # 
# peoples3.xs('전국', level=0)
# peoples2.loc['전국'].index
peoples3.index.get_level_values(0)
```

```python
peoples3.index.get_level_values(1)
```

```python
peoples3['총합'] = 0
```

```python
peoples3.loc[('전국', '전국'), '총합']
```

```python
for idx1 in peoples3.index.get_level_values(0).unique():
    for idx2 in peoples3.index.get_level_values(1).unique():
        peoples3.loc[(idx1, idx2), '총합'] = peoples3.xs(idx1, level=0).loc[idx2].sum()
```

```python
peoples3
```

```python
peoples3.loc[('전국'), '총합'].sort_values(ascending=False).head(2).index[1]
```

```python
for idx1 in peoples3.index.get_level_values(0).unique():
    if idx1 == '전국':
        continue
    print(f'{idx1}로 가장 많이 전입해온 도시: {peoples3.loc[(idx1), "총합"].sort_values(ascending=False).index[1]:>12}')
```
