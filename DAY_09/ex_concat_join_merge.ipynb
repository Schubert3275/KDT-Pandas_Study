## [데이터 병합/연결/조인]
- 여러 개의 Series/DataFrame의 객체의 데이터들을 합치기 방법

(1) 모듈 로딩 <hr>

```python
import pandas as pd
```

(2) 데이터 준비 <hr>

```python
datas = {'name': ["홍 길동", "이 나영", "마 징가", "베 토벤"],
         'age': [10, 21, 73, 89],
         'reg_date': ["2010/1/23", "2020/11/8", "1999/5/4", "1987/12/24"]}

datas2 = [["홍 길동", "이 나영", "마 징가", "베 토벤"],
          [10, 21, 73, 89],
          ["2010/1/23", "2020/11/8", "1999/5/4", "1987/12/24"]]
```

(3) 데이터 저장 <hr>

```python
# Dict ==> DataFrame
df1 = pd.DataFrame(datas)
df1
```

```python
# List ==> DataFrame
df2 = pd.DataFrame(datas2)
df2
```

(4) 데이터 정보 확인 <hr>

```python
# 데이터의 전체 기본 정보 => info()
df2.info()
```

(5) 데이터 전처리 <hr>

- (5-1) 행 <=> 열 치환

```python
df2.T
```

```python
df2 = df2.transpose()
df2.info()
```

```python
df2.head()
```

- (5-2) 데이터 타입 일치

```python
# 1번 컬럼 : object => 정수
df2[1] = df2[1].astype('uint8')
```

```python
df2.dtypes
```

```python
# 2번 컬럼 : object => datetime64[ns]
df2[2] = df2[2].astype('datetime64[ns]')
```

```python
df2.dtypes
```

- (5-3) 0번 컬럼의 이름을 성과 이름으로 분리

```python
# 0번 컬럼만 추출 => Series
nameSR = df2[0]
nameSR
```

```python
# "이 나영" 데이터를 추출 => 성과 이름으로 분리
# type(nameSR[1]), nameSR[1].upper(), nameSR[1].split()
```

```python
# for idx in range(nameSR.shape[0]):
#     print(nameSR[idx])
#     nameSR[idx] = nameSR[idx].split()
# nameSR
```

```python
namesDF = nameSR.str.split(expand=True)
```

```python
df2, namesDF
```

- (5-4) 두개의 DataFrame을 컬럼 방향으로 연결

```python
pd.concat([df2, namesDF], axis='columns', ignore_index=True)
```

- (5-5) 두개의 DataFrame을 행(row) 방향 데이터 연결

```python
# 여러 개의 DF, SR을 연결 시에 모든 컬럼명/행인덱스를 연결시켜주는 방식 ==> outer 방식(기본)
pd.concat([df2, namesDF])
```

```python
# 여러 개의 DF, SR을 연결 시에 동일 컬럼명/행인덱스만 연결시켜주는 방식 ==> inner 방식
pd.concat([df2, namesDF], join='inner')
```

### [병합 - (2) merge]
- 두 개의 DF에서 특정 컬럼을 기준으로 데이터를 합치는 것
- 두 DF에 합치는 기준이 되는 컬럼을 지정 필요
- 지정된 컬럼명이 없으면 동일한 컬럼명을 기준으로 합쳐짐

(2-1) 데이터 준비 <hr>

```python
f1 = '../DATA/stock price.xlsx'
f2 = '../DATA/stock valuation.xlsx'
```

(2-2) 데이터 저장 EXCEL ==> DataFrame <hr>

```python
DF1 = pd.read_excel(f1)
DF2 = pd.read_excel(f2)
```

(2-3) 데이터 확인 <hr>

```python
DF1.info()
```

```python
DF2.info()
```

(2-4) 데이터프레임 병합 <hr>

- (2-4-1) merge()

```python
# [기본] 두 개의 DF에서 동일 컬럼명을 찾아서 두 DF에 존재하는 컬럼명의 행 데이터만 합치기
pd.merge(DF1, DF2)
```

```python
DF1.shape, DF2.shape
```

```python
# [설정 => how 'outer'] 두 개의 DF에서 동일 컬럼명을 찾아서 모두 합치기
pd.merge(DF1, DF2, how='outer')
```

```python
# [설정 => how 'left'] 두 개의 DF에서 동일 컬럼명을 찾아서 왼쪽 DF 데이터를 기준으로 병합
pd.merge(DF1, DF2, how='left')
```

```python
# [설정 => how 'right'] 두 개의 DF에서 동일 컬럼명을 찾아서 오른쪽 DF 데이터를 기준으로 병합
pd.merge(DF1, DF2, how='right')
```

```python
# [설정 => how 'cross'] 두 개의 DF에서 동일 컬럼명을 찾아서 오른쪽 DF 데이터를 기준으로 병합
pd.merge(DF1, DF2, how='cross')
```
